```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cc_df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/car_crashes.csv')
titanic_df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv')
```

```python
cc_df.head()
```

```python
titanic_df.head()
```

```python
embarked_counts = titanic_df['Embarked'].value_counts(normalize=True)

embarked_counts.plot(kind='bar')
```

```python
embarked_counts = titanic_df['Embarked'].value_counts(normalize=True)
plt.bar(x=embarked_counts.index, height = embarked_counts)
plt.show()
```

```python
fig, ax = plt.subplots()
```

```python
fig, ax = plt.subplots(2,2)
```

```python
fig, ax = plt.subplots(2,2)
ax[0, 0].bar(x = embarked_counts.index, height = embarked_counts)
ax[1, 0].scatter(x = titanic_df['Age'], y = titanic_df['Fare'], c='orange')
plt.show()
```

```python
southampton = titanic_df.loc[titanic_df['Embarked']=='S', 'Pclass'].value_counts(normalize=True)
cherbough = titanic_df.loc[titanic_df['Embarked']=='C', 'Pclass'].value_counts(normalize=True)
queenstown = titanic_df.loc[titanic_df['Embarked']=='Q', 'Pclass'].value_counts(normalize=True)
overall = titanic_df['Pclass'].value_counts(normalize=True)
```

```python
fig, ax = plt.subplots(2, 2, figsize = (4, 4))

fig.suptitle('Comparison of Passenger Class by Embarkation')

ax[0,0].bar(x = overall.index, height = overall)
ax[0, 0].set_title( 'Passenger Classes - Overall', fontsize = 7)
ax[1,0].bar(x = southampton.index, height = southampton)
ax[1, 0].set_title( 'Passenger Classes - Southampton', fontsize = 7)
ax[0,1].bar(x = cherbough.index, height = cherbough)
ax[0, 1].set_title( 'Passenger Classes - Cherbough', fontsize = 7)
ax[1,1].bar(x = queenstown.index, height = queenstown)
ax[1, 1].set_title( 'Passenger Classes - Queenstown', fontsize = 7)

plt.tight_layout()
plt.show()
```
