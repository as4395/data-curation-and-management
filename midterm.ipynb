```python
# Part 2: Do any required data cleaning if necessary.
```

```python
df = pd.read_csv("Quality_of_Life.csv")
```

```python
df['country'] = df['country'].replace({'Us Virgin Islands': 'US Virgin Islands'})
```

```python
object_cols = df.select_dtypes(include='object').columns
df[object_cols] = df[object_cols].apply(lambda x: x.str.replace("'", "", regex=False))
```

```python
df['Quality of Life Value'] = pd.to_numeric(df['Quality of Life Value'].str.replace(':', '', regex=False), errors='coerce')
df['Property Price to Income Value'] = pd.to_numeric(df['Property Price to Income Value'].astype(str).str.replace(':', '', regex=False), errors='coerce')
```

```python
category_cols = df.columns[df.columns.str.contains('Category')]
df[category_cols] = df[category_cols].replace(['Unknown', 'None', ''], np.nan)
```

```python
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
df.dropna(subset=category_cols, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.to_csv("Quality_of_Life_Cleaned.csv", index=False)
```

```python
# Step 3: Create at least 3 plots using only the matplotlib and seaborn libraries, where one
# of each library is used
# Step 4: For each plot, a description of the findings per the visualization will be provided. 
# Each description must be at least two sentences.
```

```python
df = pd.read_csv("Quality_of_Life_Cleaned.csv")
```

```python
# Plot 1: Seaborn Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df,
    x='Quality of Life Category',
    y='Property Price to Income Value'
)
plt.title('Property Price to Income Ratio by Quality of Life Category')
plt.xlabel('Quality of Life Category')
plt.ylabel('Property Price to Income Ratio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Findings:
# Countries with a lower quality of life often show wide variation in housing affordability, with many extreme outliers. 
# In contrast, higher quality of life countries tend to have more balanced and affordable property markets.
```

```python
# Plot 2: Matplotlib Bar Chart
avg_safety = df.groupby('Quality of Life Category')['Safety Value'].mean().sort_values()
fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(avg_safety.index, avg_safety.values, color='skyblue')
ax.set_title('Average Safety Value by Quality of Life Category')
ax.set_xlabel('Quality of Life Category')
ax.set_ylabel('Average Safety Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Findings:
# Higher Quality of Life categories are clearly associated with better safety scores.
# People living in countries with a â€œVery Highâ€ quality of life tend to feel safer overall.
```

```python
# Plot 3: Seaborn Histogram with KDE
plt.figure(figsize=(12, 6))
sns.histplot(df['Quality of Life Value'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Quality of Life Values')
plt.xlabel('Quality of Life Value')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Findings:
# Most countries fall between 100 and 150 on the Quality of Life scale, with a few outliers scoring much higher.
# This suggests that while many countries enjoy decent living standards, some significantly outperform the rest.
```
