## Step 2 (1 point): Create a sqlite database.
* Name it cars.db

```python
conn = sqlite3.connect('cars.db')
```

## Step 3 (1 point): Create a table that is the cars from 2017 in the cars dataframe
* Call the table "tbl_cars"
* do not include the index in the table

#### Create a dataframe from the SQLite table to confirm that the table data is there

```python
cars_2017_df = used_car_df[used_car_df['Manufacturing_year'] == 2017]
cars_2017_df.to_sql('tbl_cars', con=conn, if_exists='replace', index=False)
test_sql = "SELECT * FROM tbl_cars"
cars_df = pd.read_sql(test_sql, con=conn)
cars_df
```

## Step 4 (2 points): Use SQLAlchemy to connect to the same cars.db database
* create another dataframe that is the cars data for 2022
* insert that data into the tbl_cars table
* read the newly updated table into a dataframe
* count the rows to confirm it matches the sum of the row counts from step 1

```python
engine = create_engine('sqlite:///cars.db')
```

```python
cars_2022_df = used_car_df[used_car_df['Manufacturing_year'] == 2022]
```

```python
cars_2022_df.to_sql('tbl_cars', con=engine, if_exists='append', index=False)
```

```python
test_sql = "SELECT * FROM tbl_cars"
updated_cars_df = pd.read_sql(test_sql, con=engine)
```

```python
print("Number of rows after inserting 2022 cars:", updated_cars_df.shape[0])
```
