```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('insurance-1.csv')

df.head()
```

### Exercise 1 (1 point):

Create a histogram using the "age" column.

```python
df.plot(y='age', kind='hist', bins = 10, title = 'Histogram of Age') 

# df["age"].plot(kind="hist", bins=10, title="Histogram of Age", xlabel="Age", ylabel="Frequency") 
# also works, but the first method better follows the lecture notes
```

### Exercise 2 (1 point):

Create a bar plot showing the count of "children" groups in the data.

```python
children_counts = df['children'].value_counts()
children_counts.plot.bar(title='Count of Children Groups', rot=0)
```

### Exercise 3 (2 points):

Create a scatter plot using the "bmi" and "charges" data.  Set the "bmi" data on the x axis and the "charges" data on the y axis.

```python
df.plot(kind='scatter', x='bmi', y='charges', title='Scatter Plot of BMI vs Charges')
```

### Exercise 4 (1 point):

Create a pie plot showing the number of smokers vs non-smokers in the data.

```python
smoker_counts = df['smoker'].value_counts()
smoker_counts.plot.pie(title='Smokers vs Non-Smokers')
```

### Bonus Exercise (1 point):

Create a horizontal bar plot on the combined group of "region" and "smoker".  Sort the plot results by "region" ascending and "smoker" descending.

```python
df.groupby(['region', 'smoker']).size().sort_index(level=['region', 'smoker'], ascending=[True, False]).plot.barh(title='Region and Smoker Group Counts')
```
